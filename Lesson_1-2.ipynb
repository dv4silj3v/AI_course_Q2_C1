```python
x = np.linspace(-5, 5, num=51)
```

```python
k1 = 1
k2 = 4
```

```python
y1 = np.cos(k1 * x)
```

```python
y2 = np.cos(k2 * x)
```

```python
plt.plot(x, y1, 'g', label='f(x)=cos(x)')
plt.plot(x, y2, 'b', label='f(x)=cos(4x)')
plt.title("Cos(x) and Cos(4x)", fontsize=16, fontweight="bold")
plt.legend()
plt.show()
```

```python
circle = plt.Circle((0.5, 0.5), 0.5, color='r')
fig, ax = plt.subplots()
ax.add_artist(circle)
```

```python
def f_circle(R, a, b): 
    # (x - a)**2 + (y - b)**2 = R**2
    plt.axis([-6, 6, 0, 6])
    x = np.linspace(-5, 5, num=51)
    y = np.sqrt(R**2 - (x - a)**2) + b
    plt.plot(x, y, 'g)
    plt.show()
```

```python
f_circle(4, 0, 0)
```

```python
def f_ellipse(a, b): 
    # x**2/a**2 + y**2/b**2 = 1, when a > b
    plt.axis([-6, 6, 0, 6])
    x = np.linspace(-5, 5, num=51)
    y = np.sqrt((1 - (x**2/a**2)) * b**2) 
    plt.plot(x, y, 'b')
    plt.show()
```

```python
f_ellipse(4, 2)
```

```python
def f_hyperb(a, b):
    # y**2/a**2 - x**2/b**2 = 1
    plt.axis([-6, 6, 0, 6])
    x = np.linspace(-5, 5, num=51)
    y = np.sqrt((1 + (x**2/b**2)) * a**2) 
    plt.plot(x, y, 'b')
    plt.show()
```

```python
f_hyperb(2, 1)
```
